```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,N0):
    return 1/(1+((1-N0)/N0)*np.exp(-x))
```

```python
data1 = np.loadtxt('data/11.txt')
data2 = np.loadtxt('data/12.txt')
data3 = np.loadtxt('data/13.txt')
data4 = np.loadtxt('data/14.txt')

data1[:,1] = (f(data1[:,0],2) - data1[:,1])/data1[:,1]
data2[:,1] = (f(data2[:,0],2) - data2[:,1])/data2[:,1]
data3[:,1] = (f(data3[:,0],2) - data3[:,1])/data3[:,1]
data4[:,1] = (f(data4[:,0],2) - data4[:,1])/data4[:,1]

print(data2.shape)
plt.plot(data1[:,0], data1[:,1], label='0.5',marker='o')
plt.plot(data2[:,0], data2[:,1], label='0.25',marker='o')
plt.plot(data3[:,0], data3[:,1], label='0.125',marker='o')
plt.plot(data4[:,0], data4[:,1], label='0.0625',marker='o')

plt.xlabel('Time')
plt.ylabel('Relative error')

plt.legend()
plt.show()
```
